```python
# define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data' 
```

```python
# Apply function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Question 2-Creating new variable "busiest days"

result2 = []

for value in df_ords_prods_merged["order_day_of_week"]:
  if value in [0,1]:
    result2.append("Busiest days")
  elif value in [3,4]:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
```

```python
df_ords_prods_merged['busiest_days'] = result2
```

```python
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

# Question 4

```python
# Determining busiest hours of day

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Busiest time of day is between 9-16, slowest time is between 0-6 and the rest is average-create new variable


result3 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value>=9 and value<=16:
    result3.append("Most orders")
  elif value>=0 and value<=6:
    result3.append("Fewest orders")
  else:
    result3.append("Average orders")
```

```python
result3
```

```python
#saving variable as a new column

df_ords_prods_merged['busiest_period_of_day'] = result3
```
